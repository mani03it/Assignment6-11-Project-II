Task: Create a sql db from adult dataset and name it sqladb

```python
import sqlite3

conn = sqlite3.connect('sqladb.db')

cur = conn.cursor()
```

```python
sql = """
    create table adult (
        age        integer,
        workclass       text,
        fnlwgt integer,
        education text,
        educationNum integer,
        maritalStatus text,
        occupation text,
        relationship text,
        race text,
        sex text,
        capitalGain integer,
        capitalLoss integer,
        hoursPerWeek integer,
        nativeCountry text,
        salary text
    );"""
cur.execute(sql)
```

	1. Select 10 records from the adult sqladb

```python
cur.execute("SELECT * FROM adult LIMIT 10;")
rows = cur.fetchall()
for row in rows:
     print(row)
```

    2. Show me the average hours per week of all men who are working in private sector

```python
cur.execute("select avg(hoursPerWeek) from adult where workclass='Private' AND sex='Male';")
rows = cur.fetchone()
print(rows)
```

	3. Show me the frequency table for education, occupation and relationship, separately	

```python
cur.execute("SELECT education, occupation,relationship, count(*) from adult GROUP by education, occupation, relationship;")
rows = cur.fetchall()
for row in rows:
     print(row)
```

    4. Are there any people who are married, working in private sector and having a master's degree

```python
cur.execute("select * from adult WHERE maritalStatus like 'Married%' AND workclass='Private' AND education='Masters';")
rows = cur.fetchall()
for row in rows:
     print(row)
```

	5. What is the average, minimum and maximum age group for people working in different sectors

```python
cur.execute("SELECT min(age) , max(age) , workclass from adult GROUP BY workclass;")
rows = cur.fetchall()
for row in rows:
     print(row)
```

	6. Calculate age distribution by country

```python
cur.execute("SELECT  nativeCountry, age ,count(*) from adult GROUP by nativeCountry,age;")
rows = cur.fetchall()
for row in rows:
     print(row)
```

	7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
cur.execute("SELECT capitalGain - capitalLoss as NetCapitalGain from adult;")
rows = cur.fetchall()
for row in rows:
     print(row)
```

	1. Create an sqlalchemy engine using a sample from the data set

```python
from numpy import genfromtxt
from time import time
from datetime import datetime
from sqlalchemy import Column, Integer, Float, Date, String
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker

def Load_Data(file_name):
    data = genfromtxt(file_name, delimiter=',', converters={0: lambda s: str(s)})
    return data.tolist()

Base = declarative_base()

class Adult_Table(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'AdultTable'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    aid = Column(Integer, primary_key=True, nullable=False) 
    age = Column(Integer) 
    workclass = Column(String(255, convert_unicode=False))
    fnlwgt = Column(Integer)
    education = Column(String(255, convert_unicode=False))
    educationNum = Column(Integer)
    maritalStatus = Column(String(255, convert_unicode=False))
    occupation = Column(String(255, convert_unicode=False))
    relationship = Column(String(255, convert_unicode=False))
    race = Column(String(255, convert_unicode=False))
    sex = Column(String(255, convert_unicode=False))
    capitalGain = Column(Integer)
    capitalLoss = Column(Integer)
    hoursPerWeek = Column(Integer)
    nativeCountry = Column(String(255, convert_unicode=False))
    salary = Column(String(255, convert_unicode=False))     
        
if __name__ == "__main__":
    t = time()

    #Create the database
    engine = create_engine('sqlite:///Adult_SQLAlchemy.db')
    Base.metadata.create_all(engine)

    #Create the session
    session = sessionmaker()
    session.configure(bind=engine)
    s = session()

    try:
        file_name = "adult.data" 
        data = Load_Data(file_name) 
        j=1
        for i in data:
            record = Adult_Table(**{
                'aid': j,
                'age' : i[0],
                'workclass' : i[1],
                'fnlwgt' : i[2],
                'education' : i[3],
                'educationNum' : i[4],
                'maritalStatus' : i[5],
                'occupation' : i[6],
                'relationship' : i[7],
                'race' : i[8],
                'sex' : i[9],
                'capitalGain' : i[10],
                'capitalLoss' : i[11],
                'hoursPerWeek' : i[12],
                'nativeCountry' : i[13],
                'salary' : i[14]                
            })
            s.add(record) #Add all the records
            j = j+1

        s.commit() #Attempt to commit all the records
    except:
        s.rollback() #Rollback the changes on error
    finally:
        s.close() #Close the connection
    print( "Time elapsed: " + str(time() - t) + " s.")
```
