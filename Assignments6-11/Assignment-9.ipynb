    1. Create a pie chart presenting the male/female proportion

```python
import pandas as pd
import matplotlib.pyplot as plt 
from pandas.plotting import table

%matplotlib inline 

url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"

titanic = pd.read_csv(url)
df = titanic["sex"].value_counts()

#To display pie chart.
ax1 = plt.subplot(121, aspect='equal')
df.plot(kind='pie', ax=ax1, autopct='%1.1f%%', startangle=90, shadow=False, legend = False, fontsize=14)

#To display the data used to draw the pie chart in the table format.
ax2 = plt.subplot(122)
plt.axis('off')
tbl = table(ax2, df, loc='right')
tbl.auto_set_font_size(False)
tbl.set_fontsize(14)
plt.show()
```
