```python
### Read Dataset
```

```python
import pandas as pd
import os
os.chdir(r"F:\Data Science Master\9-Scipy & Matplotlib\Mentor upload\Visualization Exercises")
df_csd = pd.read_csv("company_sales_data.csv")
```

```python
### Exercise 1: Read Total profit of all months and show it using a line plot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(18,5))
df_csd.plot(x= 'month_number', y='total_profit', style='-')
```

```python
### Exercise 2: Get Total profit of all months and show line plot with the following Style properties

# Line Style dotted and Line color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3
```

```python
plt.figure(figsize=(18,5))
df_csd.plot(x= 'month_number', y='total_profit',color='red', marker='o', linestyle='dotted',linewidth=3, markersize=12)
plt.legend( ('total_profit'), loc='lower right')
plt.xlabel('Month number')
plt.ylabel('Sold units number')
```

```python
### Exercise 3: Read all product sales data and show it using a multiline plot
```

```python
df_csd.plot(x="month_number", y=["facecream", "facewash","toothpaste","bathingsoap","shampoo","moisturizer"])
plt.show()
```

```python
### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
```

```python
plt.figure(figsize=(18,5))
plt.scatter(df_csd.month_number, df_csd.toothpaste, marker='o')
```

```python
### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
```

```python
ax = plt.subplot(111)
ax.bar(df_csd.month_number, df_csd.facecream, width=0.2, color='g',align='center')
ax.bar(df_csd.month_number, df_csd.facewash, width=0.4, color='r',align='center')
plt.show()
```

```python
### Exercise 6: Read the total profit of each month and show it using the histogram to see most common profit ranges
```

```python
plt.style.use('seaborn-white')
plt.hist(df_csd.month_number, weights=df_csd.total_profit,edgecolor='k')
```

```python
### Exercise 7: Calculate total sale data for last year for each product and show it using a Pie chart
```

```python
pieLabels = 'facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'
populationShare = [df_csd['facecream'].sum(),df_csd['facewash'].sum(),df_csd['toothpaste'].sum(),df_csd['bathingsoap'].sum(),df_csd['shampoo'].sum(),df_csd['moisturizer'].sum()]
figureObject, axesObject = plt.subplots()
axesObject.pie(populationShare,labels=pieLabels,autopct='%1.2f',startangle=90)
axesObject.axis('equal')
plt.show()
```

```python
### Exercise 8: Read Bathing soap facewash of all months and display it using the Subplot
```

```python
plt.figure(figsize=(18,5))
ax = plt.subplot(221)
ax.bar(df_csd.month_number, df_csd.bathingsoap, width=0.4, color='g',align='center')
ax = plt.subplot(222)
ax.bar(df_csd.month_number, df_csd.facewash, width=0.4, color='r',align='center')
plt.show()
```
